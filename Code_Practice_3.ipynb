```python
url1 = "https://raw.githubusercontent.com/fivethirtyeight/data/master/"
url2 = "college-majors/recent-grads.csv"
url = url1 + url2

df538 = pd.read_csv(url)

df538.shape
#the dimensions of df538 are 173 rows and 21 columns
```

**Part b**

```python
#to read the first 10 lines of the file:
df538.head(10)
```

**Part c**

```python
list(df538)[2],list(df538)[4],list(df538)[15],list(df538)[16],list(df538)[17]
#These variables represent 
```

* "Major" is the description of the Major/program of study
* "Men" is the number of male graduates in the respective major
* "Median" is the median earnings of a full-time, year-round worker who graduated from the respective major
* "P25th" is the 25th percentile of earnings for graduates of a respective major
* "P75th" is the 75th percentile of earnings for graduates of a respective major

**Part d**

```python
df538.set_index("Major")
```

**Part e**

```python
df538.sort_values(by=["Total"])
```

**Part f**

```python
#10 majors with the most number of people:
df538.sort_values(by="Total", ascending=False).head(10)
```

**Part g**

```python
fig, ax = plt.subplots()
df538.set_index("Major_code").sort_values(by="Median").head(10).plot(ax = ax,
                                                                     kind="barh", y="Median")

ax.set_title("Sorted by Median Salary", fontsize = 14, fontweight = "bold")
ax.set_ylabel("Major Code", fontsize = 14)
ax.set_xlabel("Median Salary", fontsize = 14)
```

```python
fig, ax = plt.subplots()
df538.set_index("Major_code").sort_values(by="P25th").head(10).plot(ax = ax,
                                                                    kind="barh", y="P25th")

ax.set_title("Sorted by 25th Percentile Salary", fontsize = 14, fontweight = "bold")
ax.set_ylabel("Major Code", fontsize = 14)
ax.set_xlabel("25th Percentile Salary", fontsize = 14)
```
